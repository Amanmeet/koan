## Classification Approach

Let's reproducing the multi-class classification of the paper using Random Forest

```julia
df = readtable("data/WISDM_ar_v1.1_transformed.txt",separator=',');
```

```julia
# combine upstairs/downstair as stairs as in the paper
for i in 1:nrow(df)
    if df[i,:class] == "Downstairs" || df[i,:class] == "Upstairs"
        df[i,:class] = "Stairs"
    end
end
```

```julia
# 80/20 split for train and test
train, test = splitrandom(df, 0.80);
```

```julia
inputrange = 2:45

train_features = array(train[:,inputrange])
train_labels = array(train[:,:class])
t = build_forest(train_labels, train_features, 10, 20, 0.5)
```

```julia
preds = apply_forest(t, array(test[:,inputrange]))
mean(preds .== test[:class])
```

```julia
cm = confusion_matrix(array(test[:class]),preds)
```

## Hidden Markov Model with Digital Signal Processing

Use a Fast Fourier Transform (FFT) to estimate the fundamental frequency of each frame then a four-state Hidden Markov Model to classify the observed sequence.

```julia
# read records for just user 33 from the top of the file
df = readtable("data/WISDM_ar_v1.1_raw.txt",separator=',', header = false, nrows=8489,
               names=[:user,:activity,:timestamp,:x,:y,:z]);
```

### Fundamental Frequency Estimation

```julia
# specify the sampling rate at 20Hz, set window size to 10 seconds 
samp_rate = 20
window_size = 10

# split dataframe into 10 second windows with no overlaps
windows = arraysplit(df[:z], samp_rate*window_size, 0);
```

```julia
# function to estimate the fundamental frequency of the frame
function estimate_ff(frame)
            
    sample_freq = fftfreq(windows.n,samp_rate)
    sig_fft = fft(frame)
    pidxs = sample_freq .> 0
    freqs = sample_freq[find(sample_freq .> 0)]
    power = abs(sig_fft)[pidxs]
    fundfreq = freqs[indmax(power)]
        
end
```

```julia
# append fundamental frequency to the dataframe
df[:fundfreq] = 0.
j = 1

for i in 1:window_size*samp_rate:nrow(df)-window_size*samp_rate
    df[i:i+window_size*samp_rate,:fundfreq] = estimate_ff(windows[j])
    j += 1
end
```

```julia
summarystats(df[:fundfreq])
```

# Hidden Markov Model (HMM)

```julia
# Hidden Markov Model code from https://gist.github.com/sbos/7264118
include("src/hmm.jl")
```

```julia
using Distributions, HMM
```

```julia
hmm = fit(HiddenMarkovModel{Normal, 4}, array(df[:fundfreq])+1e-06);
```

```julia
df[:est_z] = viterbi(hmm, array(df[:fundfreq])+1e-06);
```

```julia
summarystats(df[:est_z])
```

```julia
p1 = plot(df, x=:timestamp, y=:z, color=:activity, Geom.line, Guide.ylabel("Accel Z"), Guide.colorkey("Activity"))
p2 = plot(df, x=:timestamp, y=:z, color=:fundfreq, Geom.line, Guide.ylabel("Freq."), Guide.colorkey("Frequency"))
p3 = plot(df, x=:timestamp, y=:z, color=:est_z, Geom.line, Guide.ylabel("HMM"), Guide.colorkey("Hidden State"))

vstack(p1,p2,p3)

#draw(SVG("wisdm.svg", 6inch, 6inch), vstack(p1,p2,p3))
```
